```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def check_adult(self):
    print(f"年齢：{self.age}")
    if self.age >= 20:
      print(f"{self.name}は大人である")
    else:
      print(f"{self.name}は大人でない")
hms = [Human("名前太郎", 29), Human("名前次郎", 10), Human("名前花子", 40), Human("名前三郎", 19), Human("名前史郎", 20)]
for hm in hms:
  hm.check_adult()
```
